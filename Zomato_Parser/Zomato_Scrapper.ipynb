```python
#Import the Request and BeautifulSoup Librarys
import requests
from bs4 import BeautifulSoup

#Use Headers or else Zomato will block the access
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
r = requests.get("https://www.zomato.com/hyderabad/great-food-no-bull", headers=headers)
c=r.content

soup=BeautifulSoup(c,"html.parser")
#print(soup.prettify()) You can use this command to see the HTML code in the output window.

#Inspect and find the Class which contains all the data you need and Save it all
all = soup.find_all("div",{"class":"ptop0 pbot0 pl10 pr10"})
#type(all) -- Data is saved in bs4.element.ResultSet, ResultSet are not a list.

# for item in all:
#     print(item.find("div",{"class":"res_title zblack bold nowrap"}).text.replace("\n","").replace(" ",""))
#     print(item.find_all("div",{"class":"nowrap grey-text fontsize5 ttupper"})[0].get_text(strip=True))
#     print(item.find_all("div",{"class":"nowrap grey-text"})[0].get_text(strip=True))
#     print("\n")
```

```python
#saving the Whoel data in a list as JSON with the help of function this time,Yet Another way of doing the same thing as above.
l=[]

import json

def get_restaurant_info(soup):
    d = {}
    try:
        d["Name"]=soup.find("div",{"class":"res_title"}).text.replace("\n","").replace(" ","")
    except:
        d["Name"]= None
        pass
    
    try:
        d["Address"]=soup.find_all("div",{"class":"nowrap grey-text fontsize5 ttupper"})[0].get_text(strip=True)
    except:
        d["Address"]= None
        pass
    
    try:
        d["Food type"]=soup.find_all("div",{"class":"nowrap grey-text"})[0].get_text(strip=True)
    except:
        d["Food type"]= None 
        pass

    return d

l = [get_restaurant_info(item) for item in all]

#Below line of codes can used to save the Data in file in JSON format.
# with open('restaurants_zomato.json', 'w') as write_file:
#     json.dump(l, write_file)

#ignore this.
# from pprint import pprint
# pprint(l)
        
```
